```python
#import libraries
import numpy as np 
import matplotlib.pyplot as plt 

N = 1e5
x = np.arange(1, N+1)
hplot = np.zeros_like(x, dtype=float)
log = np.log(x)

series=0

for i in range(0, int(N)):
    series += 1/x[i]
    hplot[i] = series

plt.semilogx(x, hplot, x, log)
plt.grid()
plt.show()

#plot to show that the difference approaches eulers constant 
plt.semilogx(x, hplot-log)
plt.grid()
plt.show()
```

```python
#Siri's triangle changed to mod 4
n = 30
mod = 4


def binom(n, k):
    a = 1
    for i in range(k):
        a = a*(n-i)//(i+1)
    return a


def symb(u):
    if u == 0:
        return "\u001b[33m0"
    elif u==1:
        return "\u001b[34m1"
    elif u == 2:
        return "\u001b[31m2"
    else:
        return "\u001b[32m3"


for i in range(n+1):
    for j in range(0, n-i+1):
        print(' ', end='')
    for j in range(i+1):
        u = binom(i, j) % mod
        s = symb(u)
        print(' ', s, sep='', end='')
    print('')
```
