```python
#import libraries 
import numpy as np
import matplotlib.pyplot as plt 


#function to arrange numbers in an array into the winding pattern
def winding_pattern(num):
    #find next square number after num and the size of the array
    next_square = int(np.ceil(np.sqrt(num))**2)
    m = int(np.sqrt(next_square))
    array = [[0] * m for i in range(m)]

    #find the number of rows and columns
    row, col = m // 2, m // 2

    #begin plotting the numbers in the array, placing 1 in the middle 
    array[m // 2][m // 2] = 1
    next_num = 2

    step = 1
    while next_num <= num:
        #place number to the right of the previous
        for i in range(step if step % 2 != 0 else step - 1):
            if col < m - 1:
                col += 1
                array[row][col] = next_num
                next_num += 1
                if next_num > num:
                    break
        if next_num > num:
            break
        #place number above the previous
        for i in range(step):
            if row > 0:
                row -= 1
                array[row][col] = next_num
                next_num += 1
                if next_num > num:
                    break
        if next_num > num:
            break
        step += 1

        #place number to the left of the previous
        for i in range(step):
            if col > 0:
                col -= 1
                array[row][col] = next_num
                next_num += 1
                if next_num > num:
                    break
        if next_num > num:
            break
        #place number below the previous
        for i in range(step):
            if row < m - 1:
                row += 1
                array[row][col] = next_num
                next_num += 1
                if next_num > num:
                    break
        if next_num > num:
            break
        step += 1
    return array

#pick largest desired number in the grid and calculate the grid. 
num = 100000
array = winding_pattern(num)


#highlight multiples of k
#k = 5
#for row in range(len(array)):
    #for col in range(len(array[row])):
        #if array[row][col] % k == 0:
            #array[row][col] = 0

#highlight powers of 2
##for i in range(len(array)):
    #for j in range(len(array[i])):
        ##if array[i][j] != 0 and (array[i][j] & (array[i][j] - 1)) == 0:
            #array[i][j] = 0

#function to return True if n is prime 
def prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

#highlight the prime numbers
for i in range(len(array)):
    for j in range(len(array[i])):
        if prime(array[i][j]) == True:
            array[i][j] = 0


#plot using plt.imshow
labels = False
if labels:
    for ind, x in np.ndenumerate(array):
        plt.text(s=str(x), x=ind[1], y=ind[0], va='center', ha='center')

fig = plt.figure(figsize=(19,19))
plt.imshow(array, cmap='gray', vmin=0, vmax=1)
plt.axis(False)
plt.show()
```

```python
#function to find next square number
def next_square(N):
    return int(np.ceil(np.sqrt(N))**2)

#function to find position of any number N
def position(N):
    #find the difference between N and next square, and the parity of the next square 
    diff = next_square(N) - N 
    root_diff = diff - np.sqrt(next_square(N))
    if next_square(N) % 2 == 0:
        parity = 'even'
    else:
        parity = 'odd'
    if parity == 'even':
        if diff <= np.sqrt(next_square(N)):
            print(str(N) + ' is ' + str(diff) + ' positions along the top row of the ' +
                  str(np.sqrt(next_square(N))), 'by ' + str(np.sqrt(next_square(N))) + ' grid.')
        else:
            print(str(N) + ' is ' + str(root_diff) + ' positions down the rightmost column of the ' +
                  str(np.sqrt(next_square(N))), 'by ' + str(np.sqrt(next_square(N))) + ' grid.')
    else:
        if diff <= np.sqrt(next_square(N)):
            print(str(N) + ' is ' + str(diff) + ' positions along the bottom row of the ' +
                  str(np.sqrt(next_square(N))), 'by ' + str(np.sqrt(next_square(N))) + ' grid.')
        else:
            print(str(N) + ' is ' + str(root_diff) + ' positions up from the leftmost row of the ' + 
                str(np.sqrt(next_square(N))), 'by ' + str(np.sqrt(next_square(N))) + ' grid.')


#find position of 10 
position(10)
```
